```python
import cv2
import numpy as np
image=cv2.imread('C:\\Users\\Manish\\Desktop\\shape\\tir5.png')
img = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2.imshow('original',image)
cv2.waitKey(0)
# cv2.imshow('gray',img)
blur=cv2.blur(img,(7,7))
gauss = cv2.GaussianBlur(img,(7,7),0)
median = cv2.medianBlur(img,5)
bilat = cv2.bilateralFilter(img,9,75,75)
# kernel = np.ones((5,5),np.float32)/25
# blur = cv2.filter2D(img,-3,kernel)
# cv2.imshow('blur',blur)
# cv2.imshow('gauss',gauss)
# cv2.imshow('median',median)
# cv2.imshow('bilateral',bilat)
edged = cv2.Canny(img, 50,200)
# print(edged)
# cv2.imshow('canny1',edged)
contours, hierarchy = cv2.findContours(edged, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
print(contours)
print(hierarchy)
# cv2.imshow('after contour',edged)
cv2.drawContours(image, contours, -1, (255,0,0), 3)

print("Number of Contours found = " + str(len(contours)))
for cnt in contours:
    approx = cv2.approxPolyDP(cnt, 0.01*cv2.arcLength(cnt,True),True)
    if(len(approx)==3):
        cv2.putText(image,'Triangle',(50,160),cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,0),2)
        cv2.imshow('Contours', image)
    elif(len(approx)==4):
        cv2.putText(image,'Square',(50,175),cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,0),2)
        cv2.imshow('Contours', image)
    elif(len(approx)>=15):
        cv2.putText(image,'Circle',(60,170),cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,0),2)
        cv2.imshow('Contours', image)
        
# cv2.imshow('Contours', image)
print(len(approx))
# laplacian = cv2.Laplacian(median,cv2.CV_64F)
# sobelx = cv2.Sobel(median,cv2.CV_64F,1,0,ksize=5)
# sobely = cv2.Sobel(median,cv2.CV_64F,0,1,ksize=5)
# cv2.imshow('laplacian',laplacian)
# cv2.imshow('sobelx',sobelx)
# cv2.imshow('sobely',sobely)

# print(img.shape)
# print(img)
print(img.shape)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
